                                                Git konfigurálása

Elsőnek is tudatnunk kell a Git-el, hogy kik is vagyunk.

Ehhez kelleni fog a "Saját nevünk" és e-mail címünk.

```python
!git config --global user.name "Random Bob"
!git config --global user.email "random.bob@example.com"
```

Git adatbázist egyszerű létrehozni(a kívánt könyvtárban adjuk ki a parancsot)

```python
!git init
```

Vagy, ha esetleg, másét szeretnénk beszerezni, az se okoz gondot.

```python
!git clone user@example.server.com:proba.git
```

Attól még, hogy létrehoztunk pár file-t a könyvtárban,azok nem fognak a gitbe felkerülni, ugyanis elsőnek ki kell jelőlnünk, hogy mely fájlokat szeretnénk feltölteni.

```python
!git add file1 file2 file3
```

Ha netán az összeset felszeretnénk:

```python
!git add .
```

Ezek után, ki kell adni ezt a parancsot:

```python
!git commit 
```

Ilyenkor megfog jelenni egy editor, amely segítségével tudsz írni egy rövid leírást, hogy mit is csináltál.

Most a megváltozott adatbázisunkat szeretnénk visszatölteni a Szerverre. Ha klónozással szereztük az adatbázist, akkor ehhez csak annyit kell mondanunk, hogy:

```python
!git push
```

Ha viszont a proba könyvtárunkról az adatbázis újonan készült, akkor a Szerveren a projfelh felhasználó üzemeltetőjének el kell küldeni emailben a proba nevet, hogy tudathassa a Szerverrel, hogy lesz egy ilyen új adatbázis rajta. Ha tudatta vele, akkor a következőket kell tenni:

```python
!git remote add origin user@example.server.com:proba.git 
```

Ezután pedigami szabad fordításban annyit tesz, hogy légy szíves az előbb definiált origin nevű központi helyre töltsd fel az adatbázisunkat. A master egyelőre az egyetlen ága a központi adatbázisnak, de ezt is be kell írni.


```python
!git push origin master
```

Ha ezután valamikor máskor dolgozni szeretnénk a könyvtárban, akkor:

```python
!git pull origin master
```

Van egy biztosabb módszer is rá:

```python
!git fetch
!git diff master..origin/master
```

Ez a verzió nem írja át a könyvtárunkban a fileokat, de megmutatja mi a különbség a Szerver verziója és a mi verziónk között. És ha minden passzol akkor adjuk ezt a parancsot. 

```python
!git merge master origin/master 
```
